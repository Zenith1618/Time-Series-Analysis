##Time Resampling

```python
# A: Year end frequency
df_tesla.resample(rule='A').min()   #This is telling the open, high, low.... of the particular year
```

```python
df_tesla.resample(rule='A').max()
```

```python
# So if we just plot this data
df_tesla.resample(rule='A').min()['Open'].plot()
```

```python
# Quartely start frequency
df_tesla.resample(rule='QS').max()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
# Business End Frequency
df_tesla.resample(rule='BA').max()
```

```python
df_tesla.resample(rule='BA').max()['High'].plot()
```

```python
# Business Quarter
df_tesla.resample(rule='BQS').max()
```

```python
# Plotting
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
# Plotting Monthly max
df_tesla['Open'].resample(rule='M').max().plot(kind='bar', figsize=(15,6))
```

```python
df_tesla.head()
```

```python
# Rolling(taking the window)
#You can use different types of AGG functions
df_tesla['High'].rolling(10).mean().head(20)

# axis = 0 means we are doing it for the rows
```

```python
# Making new column
df_tesla['Open: 30 days rolling'] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head(30)
```

```python
# Plot open and rolling 30 days open
df_tesla[['Open', 'Open: 30 days rolling']].plot(figsize=(12,5))


# using the rolling even smoothing is also possible
```

# Assignment

Read the Microsoft Data using Pandas Data reader

Get the maximum price of the share from 2017 to 2022

Which is the date of the highest price of the stock?

Which is the date of the lowest price of the stock?


```python
df = pdr.DataReader('MSFT', 'stooq')
```

```python
df.head()
```

```python
df.loc['2018-01-01':'2021-01-01']['High'].resample(rule='A').max()
```

```python
df[df['High'].max() == df['High']]
```

```python
df[df['High'].max() == df['High']].index
```

```python
# Or the ChatGPT way, where idxmax basically gives the index of the max, works similar with idxmin

max_price_date = df['High'].idxmax()
max_price = df.loc[max_price_date, 'High']

print(max_price_date, max_price)
```

```python
df[df['Low'].min() == df['Low']]
```

```python
df[df['Low'].min() == df['Low']].index
```

```python
# OR

min_price_date = df['Low'].idxmin()
min_price = df.loc[min_price_date, 'Low']

print(min_price_date, min_price)
```
