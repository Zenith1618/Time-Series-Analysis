```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla = pdr.DataReader('TSLA', 'stooq')
```

```python
df_tesla
```

**Simple Moving Average(SMA):** To do smoothening the garph

>

>





```python
df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Open: 10 days rolling'] = df_tesla['Open'].rolling(window = 10, min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling']].plot(xlim = ['2021-01-01','2022-01-01'],figsize=(15,6))
```

```python
df_tesla['Open: 30 days rolling'] = df_tesla['Open'].rolling(window = 30, min_periods=1).mean()
df_tesla['Open: 50 days rolling'] = df_tesla['Open'].rolling(window = 50, min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling','Open: 30 days rolling','Open: 50 days rolling']].plot(xlim = ['2021-01-01','2022-01-01'],figsize=(15,6))
```

**Cumulative Moving Average(CMA)**

```python
# Using Function Expanding

df_tesla['Open'].expanding().mean().plot(figsize=(15,6))
```

**Exponential Moving Average(EMA)**

```python
# Let the smoothening factor(alpha: hyperparameter) be 0.1

df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1']].plot(xlim = ['2021-01-01','2022-01-01'],figsize=(15,6))
```

```python
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(alpha=0.3, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1', 'EMA_0.3']].plot(xlim = ['2021-01-01','2022-01-01'],figsize=(15,6))
```

**Exponential Weighted Moving Average(EWMA)**: Check Notebook

```python
df_tesla['EMA_5days'] = df_tesla['Open'].ewm(span=5, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1', 'EMA_5days']].plot(figsize=(15,6))
```
